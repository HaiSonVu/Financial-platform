# Data Download

Download adjusted Dow 30 price data and save the close prices used by the factor notebooks.

```python
import matplotlib.pyplot as plt
import pandas as pd
import yfinance as yf
```

```python
tickers = [
    "AAPL", "AMGN", "AMZN", "AXP", "BA",
    "CAT", "CRM", "CSCO", "CVX", "DIS",
    "GS", "HD", "HON", "IBM", "JNJ",
    "JPM", "KO", "MCD", "MMM", "MRK",
    "MSFT", "NKE", "NVDA", "PG", "SHW",
    "TRV", "UNH", "V", "VZ", "WMT",
]

data = yf.download(
    tickers,
    start="2018-01-01",
    end="2025-01-01",
    auto_adjust=True,
)

data.head()
```

```python
prices = data["Close"]

prices.head()
```

```python
prices.plot(
    figsize=(12, 6),
    title="Stock Prices",
)
plt.show()
```

```python
prices.to_csv("../data/prices.csv")
```
